First, let's define a simple couple of vectors `x` and `y`.

```matlab
x = linspace(-2*pi,2*pi,40)
```

```matlab
y = sin(x)
```

And let's make a plot!

```matlab
plot(x,y,'-*')
xlabel('x'); ylabel('y = sin(x)');
title('Test of my function')
```

The above is a plot of the function $f(x) = \sin x$, with $x \in [-\pi,2\pi]$.
