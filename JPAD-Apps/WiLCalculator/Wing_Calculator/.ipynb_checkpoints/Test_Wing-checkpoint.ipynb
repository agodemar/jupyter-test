# Introduction

In order to evaluate the wing lift curve it is necessary to obtain the following elements:
-  Zero lift angle, $\alpha_{0L}$
-  Lift coefficient slope, $C_{L\alpha}$
-  End of linearity angle, $\alpha^{\star}$
-  Maximum lift coefficient, $C_{L,\mathrm{max}}$
-  Stall angle of Attack, $\alpha_{\mathrm{s}}$


# Methods

The elements introduced in the previous section, useful to define the complete lift curve of the wing, are obtained using the following methods.
- $\alpha_{0L}$, $C_{L\alpha}$ $\longrightarrow$ Nasa-Blackwell method (see ref. XX)
- $\alpha^{\star}$ $\longrightarrow$ Mean Arifoil 
- $C_{L,\mathrm{max}}$ $\longrightarrow$ Stall path
- $\alpha_{\mathrm{s}}$ $\longrightarrow$ Semi-empirical charts

The *Nasa-Blackwell method* is a numerical method for calculating the subsonic load distribution for arbitrary lifting surface arrangements at a fixed angle of attack. The method is suitable for swept wings, non-planar wings, wing with pylons and/or end-plates; it can be also used to study the aerodynamic interaction of the wing and the horizontal tail. The resutls of this calculation method are in good agreement with the experimental results. The lifting surfaces are divided in several rectangular horse-shoe vortices along the span; one horseshoe vortex along the chord is used, that is, the midpoints of the vortices are placed only at points along the quarter-chord lines.

**`TODO:`** insert figure with influence areas.

An equal number of control points are located along the three-quarter-chord lines. The velocity from the total vortex system is equated to the component of free-stream velocity normal to the lifting surface chord at each control point. Application of this tangent-flow boundary condition for a symmetrical loading provides a set of N simultaneous equations in the N unknown circulation strengths. Solution of this set of equations provides the loading distributions over the lifting surfaces. Mach number effect is introduced through a Prandtl-Glauert correction.

The *mean airfoil* is an airfoil that has the mean characteristics of the wing airfoils calculated using influence areas. 

The *Nasa-Blackwell method* is an inviscid method for calculating the subsonic aerodynamic load distributions for arbitrary lifting-surface arrangements. This means that it does not see the non linear trait of the lift curve. So in correspondence of the calculated value of $C_{L,\mathrm{max}}$, the angle of attack obtained from the Nasa-Blackwell method is not correct because it is the maximum angle of attack that you would get if you reach the maximum $C_L$ linearly. So first it is evaluated the angle of attack at maximum lift coefficient through the linear trend of the lift line, then it’s added to this angle the increment evaluated by the diagram in the following figure, this is valid strictly for wing with high taper ratio without twist, with unique airfoil type and Mach number included between 0.2 and 0.6.
